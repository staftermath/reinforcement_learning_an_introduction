# Simulate 10-armed testbed

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, "/home/weiwen/Documents/study/reinforcement_learning/")
import warnings
warnings.filterwarnings("ignore")
```

```python
from multi_armed_bandits.multi_armed_bandits import Bandit as MyBandit
```

```python
n_actions = 10
n_steps = 1000
n_samples = 2000
epsilons = [0, 0.01, 0.1]
colors = ['r', 'g', 'b']
select_first = 3
fig, axs = plt.subplots(1,2, figsize=(16, 8))

for i, epsilon in enumerate(epsilons[:select_first]):
    label = "epsilon={:.2f}".format(epsilon)
    bandit = MyBandit(epsilon=epsilon, n_bandits=n_actions, n_steps=n_steps, n_simulations=n_samples)
    rewards, optimum_actions = bandit.simulate()
    fig.suptitle('Vertically stacked subplots')    
    axs[0].plot(range(n_steps), rewards, color=colors[:select_first][i], label=label)
    axs[1].plot(range(n_steps), optimum_actions, color=colors[:select_first][i], label=label)
axs[0].legend()
axs[1].legend()
plt.show()
```

# With Initial

```python
n_actions = 10
n_steps = 1000
n_samples = 2000
epsilons = [0.1, 0]
colors = ['r', 'g']
initial = [0, 5]
fig, axs = plt.subplots(1,2, figsize=(16, 8))

for i, epsilon in enumerate(epsilons):
    label = "epsilon={:.2f}".format(epsilon)
    bandit = MyBandit(epsilon=epsilon, n_bandits=n_actions, n_steps=n_steps, 
                      n_simulations=n_samples, initial=initial[i], step_size=0.1)
    rewards, optimum_actions = bandit.simulate()
    fig.suptitle('Initial')    
    axs[0].plot(range(n_steps), rewards, color=colors[i], label=label)
    axs[1].plot(range(n_steps), optimum_actions, color=colors[i], label=label)
axs[0].legend()
axs[1].legend()
plt.show()
```

# UCB

```python
n_actions = 10
n_steps = 1000
n_samples = 2000
epsilons = [0.1, 0]
colors = ['r', 'g']
ucb = [None, 2.]
fig, axs = plt.subplots(1,2, figsize=(16, 8))

for i, epsilon in enumerate(epsilons):
    label = "epsilon={:.2f}".format(epsilon)
    bandit = MyBandit(epsilon=epsilon, n_bandits=n_actions, n_steps=n_steps, 
                      n_simulations=n_samples, ucb=ucb[i])
    rewards, optimum_actions = bandit.simulate()
    fig.suptitle('Initial')    
    axs[0].plot(range(n_steps), rewards, color=colors[i], label=label)
    axs[1].plot(range(n_steps), optimum_actions, color=colors[i], label=label)
axs[0].legend()
axs[1].legend()
plt.show()
```
